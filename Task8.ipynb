```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

```

```python
# Load dataset
df = pd.read_csv('placement.csv')
df = df[['cgpa', 'package']]
print("First 5 row:")
print(df.head())
```

# 1.Box and whisker plot

```python

plt.figure(figsize=(10, 5))
sns.boxplot(data=df)
plt.title("Box and Whisker Plot - Sepal Dimensions")
plt.show()

```

```python
# Calculate Quartiles and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

print("\nQuartiles and IQR:")
print(f"Q1:\n{Q1}")
print(f"Q3:\n{Q3}")
print(f"IQR:\n{IQR}")

```

```python
# Tukey's Method for Outliers
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
print("\nOutliers using Tukey's Method:")
print(df[outliers.any(axis=1)])
```

# 2.z score

```python
# Calculate Z-Scores
z_scores = zscore(df)
z_df = pd.DataFrame(z_scores, columns=df.columns)

# Threshold for outlier
threshold = 3
z_outliers = (np.abs(z_df) > threshold)

print("\nOutliers using Z-Score Method:")
print(df[z_outliers.any(axis=1)])
```

# 3.scatter plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='cgpa', y='package', data=df)
plt.title("Scatter Plot - cgpa vs package")
plt.xlabel("cpga")
plt.ylabel("package")
plt.grid(True)
plt.show()
```
